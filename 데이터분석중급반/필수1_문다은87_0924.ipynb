## 특이도 Specificity
- 특이도 = TN / (TN + FP )
- 거짓양성과 참음성으로 분류되 개체 중 참음성의 비율
e.g.) 실제로 심장질환이 없는 사람 중에 올바르게 분류된 사람들의 분류
- 특이도가 높은 검사법은 양성이 나올 경우 FP 가능성이 낮다. 그래서 양성 값이 나오면 더 신뢰할 수 있다.


# G- mean 기하평균
- 루트 (specificity * recall)
- 모델의 양성 및 음성 클래스에 대한 평균 예측 성능
- 불균형 데이터에서 모델의 성능을 평가하는데 유용
불균형 데이터에서 일반적인 평가지표인 정확도가 왜곡될 수 있는 경우를 보완하고, 모델이 양성 및 음성 클래스를 모두 골고루 잘 예측하는 성능을 측정할 수 있도록 도와준다.
- 모델의 예측능력과 신뢰성을 동시에 평가할 수 있는 지표로서 활용
- 1에 가까운 값을 가질수록 모델의 성능이 높다고 해석 가능
