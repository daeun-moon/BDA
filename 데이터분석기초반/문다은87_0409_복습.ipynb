```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 막대그래프
plt.bar(range(5),[10,20,30,40,50])
```

```python
#눈금값 설정
plt.bar(range(5),[10,20,30,40,50])
plt.xticks(range(5),['a','b','c','d','e'])
```

```python
# style 설정
plt.bar(range(5),[10,20,30,40,50],color = 'red',width = 0.6, linewidth =2)
plt.xticks(range(5),['a','b','c','d','e'])
```

```python
# 색상 각각 다르게 설정
plt.bar(range(5),[10,20,30,40,50],color = ['red','blue','orange','green','purple'],width = 0.6,linewidth=2 )
plt.xticks(range(5),['a','b','c','d','e'])
```

```python
# box plot -> 이상치를 한 눈에 볼 수있다
x =np.random.normal(2,5,100)
y = np.random.normal(1,6,100)
plt.boxplot(x)
```

```python
# histogram : 도수분포표를 시각화한것
x =np.random.normal(5,10,100)
plt.hist(x)
```

```python
# 히트맵 : 2차원 배열로 각 요소의 색상을 시각화하는 것
# 상관계수 표현할 때와 비슷
data = pd.DataFrame({'DA':[50,70,80],'PG':[10,20,30]},index = ['A','B','C'])
data
```

```python
sns.heatmap(data)
```

```python
sns.heatmap(data, annot = True, fmt='d')
# annot : 박스 대응해서 데이터 값 표시, fmt는 정수라는 뜻
```

```python
# linewidth : 히트맵 선으로 쪼개지게 만든다. 색상도 지정가능
sns.heatmap(data, annot =True, fmt = 'd',linewidth=10, linecolor='white')
```

```python
#cmap으로 컬러스타일도 지정가능
sns.heatmap(data,annot=True,fmt='d',linewidth=10,linecolor='white',cmap='Greys')
```

```python
mpg =pd.read_csv('mpg.csv')
mpg
```

```python
# 잘못된 그래프 예시
mpg.plot(kind='line',x='year',y=['cty','hwy'],figsize=(10,7))
```

```python
# 다양한 시각화 그래프 여러개 그리기
# subplots 을 통해서 그래프를 여러개 그릴 수 있다.
# fig :도화지의 개념.
# 실제 그림을 그리는 것은 axis , ax를 통해서 그림을 그린다.
fig,(ax1,ax2)= plt.subplots(nrows=1, ncols=2, figsize=(10,8))
```

```python
fig,(ax1,ax2,ax3,ax4)= plt.subplots(nrows=1, ncols=4, figsize=(10,8))
```

```python
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(10,8))
# ax로만 하고 row,col 값 지정해주면 알아서 개수 매칭
ax[0].plot(mpg['cty'])
ax[1].hist(mpg['model'].value_counts(), facecolor='red')
```

```python
# 반복문 활용 다양한 그래프 그리기
#gridspec 라이브러리 활용
import matplotlib.gridspec as gridspec
grid=gridspec.GridSpec(2,2)
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace=0.5, hspace=0.4) # 여러개 그리프 충돌방지 위해 간격조절
mpg_features=['manufacturer','model','category','drv']
for idx, feature in enumerate(mpg_features):
    ax=plt.subplot(grid[idx])
    sns.countplot(x=feature, data =mpg, hue='drv',palette='pastel',ax=ax)
    ax.set_title(f'{feature}Distribution')
```
