# 판다스 str 문법
- 개념정리 & 코드 구현
- 어떤 식으로 작동하는지 주석처리로 설명

```python
# str.split()
# 괄호 안에 있는 값을를 기준으로 문자열을 나눈 다음 리스트로 묶어준다.
survey['comment'].str.split() #기본은 공백(띄어쓰기)
```

```python
# str.get()
# 인덱스로 접근하여 리스트,튜플,문자열로 부터 일부 요소를 가져온다.
survey['datetime'].str.split("-").str.get(1) # 월만 가져오기
```

```python
# str.startswidth()
# 문자열이 특정문자열로 시작하는지 확인 할 수 있다.
survey['comment'].str.startswith('역앞에')
```

```python
# str.endswidth()
# 문자열이 특정문자열로 끝나는지 확인 할 수 있다.
survey['comment'].str.endswith('좋다')
```

```python
# str.contains()
# 문자열이 특정문자열을 포함하는지 확인 할 수 있다.
survey['comment'].str.contains('초등학교')
```

```python
# str.replace()
# 특정 문자열을 다른 문자열로 대체 할 수 있다
survey['comment'].str.replace('좋다','나쁘다')
```

```python
# str.extract()
# (정규표현식)을 활용하여 해당 정규표현식에 맞는 부분만 추출 할 수 있다
survey['comment'].str.extract(pat = '(\(.+?\))')
```

```python
# str.pad()
# 문자열의 길이에 맞게 채우는 함수
# 너비가 15가 되게끔 오른쪽으로 빈 문자열을 -로 채워라
survey['comment'].str.pad(width=15, side='right', fillchar='-')
```

```python
data = [[' Choi',22],['Kim ',48],['Jo o',32]]
df = pd.DataFrame(data, columns=['Name','Age'])
df
```

```python
# str.strip()
# 인자로 전달된 문자를 String의 왼쪽과 오른쪽에서 제거합니다.
df['Name'].str.strip()
```

```python
# str.lower()
# 인자로 전달된 문자를 소문자로 바꾼다
df['Name'].str.lower()
```

```python
# str.upper()
# 인자로 전달된 문자를 대문자로 바꾼다
df['Name'].str.upper()
```

```python
# str.match
# match 메서드는 문자열의 처음부터 정규식과 매치되는지 조사
df['Name'].str.match('K')
```

```python
# str.count('')
# count 메서드는 문자열에 해당 문자열이 몇 번 들어가 있는지 빈도를 조사
df['Name'].str.count('o')
```
