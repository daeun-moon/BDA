```python
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
# 기본 형태 : scatter는 점으로 찍어지는 반면, plot은 점으로 연결된 형태를 보인다.
sns.set()
plt.scatter([1,2,3],[4,5,6])
plt.plot([1,2,3],[4,5,6],color = 'red')
```

```python
# color : 색상지정은 plot과 동일하게 color로 옵션 지정
sns.set()
plt.scatter([1,2,3],[4,5,6], color = 'red')
```

```python
# linestyle의 경우, scatter는 점을 찍기 때문에 
# 점의 style을 바꿔주는 marker 옵션을 준다. 
sns.set()
plt.scatter([1,2,3],[4,5,6], color = 'red', marker = '*')
```

```python
# linewidth의 경우, scatter는 점을 찍기 때문에 
# 점의 테두리선 굵기를 바꿔준다. 
sns.set()
plt.scatter([1,2,3],[4,5,6], color = 'red', marker = '*',linewidths =10)
```
