```python
import pandas as pd
import datetime 
crime = pd.read_csv('crime.csv')
```

```python
crime['REPORTED_DATE']=pd.to_datetime(crime['REPORTED_DATE'])
crime_dt =crime.set_index('REPORTED_DATE')
crime_dt14=crime_dt.loc['2014'].sort_index()
```

```python
# 1. first : 제일 처음 날짜를 기준으로 5개의 일들 안의 데이터 제시
crime_dt14.first('5D')
```

```python
# 2. last : 제일 끝 날짜를 기준으로 5개의 월들 안의 데이터 제시
crime_dt14.last('5M')
```

```python
# 3. weekday() : 정수로 요일을 반환. 월요일은 0이고 일요일은 6. 
crime_dt14.index[0].weekday()
```

```python
# 4. ctime() : 날짜와 시간을 나타내는 문자열을 반환
crime_dt14.index[0].ctime()
```

```python
#5 . isocalendar() : 세 개의 컴포넌트를 가진 네임드 튜플을 반환
crime_dt14.index[0].isocalendar()
```
