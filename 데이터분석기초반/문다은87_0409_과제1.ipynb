# 도수분포표
- 자료의 분포를 몇 개의 구간으로 나누고, 나누어진 각 구간에 속하는 자료가 몇 개인지 정리한 표이다

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.normal(5,10,100)
plt.hist(x)
```

```python
# 히스토그램의 경우 bins를 통해 구간 개수 설정 가능
plt.hist(x, bins=5)
```

```python
# 누적 히스토그램을 그리고 싶다면 cumulative = True로 옵션 설정
plt.hist(x,cumulative = True)
```

```python
# 히스토그램 종류 지정하기
# 기본형태인 bar 
print(plt.hist(x, histtype = 'bar'))
```

```python
# 외곽형태로 나타나는 step형태
print(plt.hist(x, histtype = 'step'))
```
