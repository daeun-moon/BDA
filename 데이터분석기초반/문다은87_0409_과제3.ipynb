```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
import seaborn as sns
titanic = sns.load_dataset('titanic') 
import warnings
warnings.filterwarnings(action  = 'ignore')
import matplotlib.pyplot as plt
```

```python
# mpg 데이터 이용 2x2 시각화
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2)
plt.figure(figsize = (16,10))
plt.subplots_adjust(wspace = 0.5, hspace = 0.4)
mpg_features = ['cyl', 'trans', 'drv','category']
for idx , feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    sns.countplot(x = feature, data = mpg, hue = 'year', palette = 'pastel',ax=ax)
    ax.set_title(f'{feature} Distribution')
    
```

```python
# titanic 데이터 이용 3x3시각화
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(3,3)
plt.figure(figsize = (16,10))
plt.subplots_adjust(wspace = 0.5, hspace = 0.4)
mpg_features = ['pclass','sibsp','parch','embarked','embark_town','alive','deck','adult_male','alone']
for idx , feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    sns.countplot(x = feature, data = titanic, hue = 'sex', palette = 'pastel',ax=ax)
    ax.set_title(f'{feature} Distribution')
```
