## 필수과제2
- mpg, 타이타닉 데이터를 groupby 등으로 깔끔하게 정리한 데이터를 비교 시각화 진행해 주세요.
- 최소 시각화는 5개 이상을 해주시고 ( 우리가 배웠던 다양한 그래프는 필수적으로 1개 이상씩은 진행해 주셔야 합니다. )


```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
import seaborn as sns
titanic = sns.load_dataset('titanic') 
import warnings
warnings.filterwarnings(action  = 'ignore')
import matplotlib.pyplot as plt
```

```python
mpg
```

```python
titanic
```

```python
#1 mpg data : fl별 cty의 평균 그래프 bar 그래프 구하기
df_1 = mpg.groupby(mpg['fl'],as_index = False)['cty'].mean()
plt.ylabel('Mean of Cty')
plt.bar(df_1['fl'], df_1['cty'])
```

```python
#2 titanic data : pclass 별 fare boxplot 그리기
df_2 = titanic[['fare']].groupby(titanic['pclass'])
df_2.boxplot(figsize = (20,10))
```

```python
#3 titanic data : sex별 age 히스토그램 그리기
df_3 = titanic[['age']].groupby(titanic['sex'])
df_3.hist(figsize = (20,10))
```

```python
#4 mpg 데이터를 cyl별로 groupby 한후 hwy의 평균을 선그래프로 작성
df_4 = mpg.groupby('cyl')['hwy'].mean()
df_4.plot(kind='line')
```

```python
#5 titanic data : parch별 age 히스토그램 그리기
df_5 = titanic[['age']].groupby(titanic['parch'])
df_5.hist(figsize = (20,10))
```
