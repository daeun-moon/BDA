```python
## 이진 분류의 경우
df_base.groupby('native-country')['class'].mean().sort_values()
# 1이 연봉이 높은 것
# 0이 연봉이 낮은 것
# 평균이 높다는 것은 -> 연봉이 높다는 것
# class의 평균이 높은 국가들은 1에 대한 비중이 높다.
```

- class가 높은 국가끼리 묶어주면
1. class가 높은 애들은 어떤 특성이 있나? 2차 분석이 필요하다. -> 아 다른 변수들도 뭔가 건들 수 있겠다.
2. unique한 값도 생각을 해서 진행해야 한다.
- 피처엔지니어링 개념으로 접근시에는 이 새로운 파생변수로 사용 할 수 있다

### class 우리가 예측해야 할 정답에 대해서 근거 기준을 가지고 꼭 데이터를 전처리 해야 한다!

# 기본 베이스 라인 모델로 모델링을 해보자!

```python
#기본 베이스 라인 모델을 모델링 해보자!
df_base=pd.get_dummies(df_base)
```

```python
df_base= df_base*1
```

```python
df_base.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_base.drop('class',axis=1),df_base['class'], test_size=0.4, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

# 파생변수 만들어서 피처엔지니어링으로 변수를 최소화하고 진행해보자

```python
df_base2= df[['age','education-num','sex','hours-per-week','native-country','race','class']]
```

```python
df_base2.dropna(inplace=True)
```

```python
df_base2
```

```python
df_fn=df_base2.groupby('native-country')['class'].mean()
```

```python
df_fn=df_fn.reset_index()
```

```python
df_base2_final=df_base2.merge(df_fn, on='native-country',how='left')
```

```python
df_base2f =df_base2_final[['age','education-num','sex','hours-per-week','class_x','class_y','race']]
```

```python
df_base2f=df_base2f.rename(columns = {'class_x':'class', 'class_y':'native-country'})
```

```python
df_base2f
```

```python
df_base2f=pd.get_dummies(df_base2f)
```

```python
df_base2f=df_base2f*1
```

```python
df_base2f
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_base2f.drop('class',axis=1),df_base2f['class'], test_size=0.4, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```
