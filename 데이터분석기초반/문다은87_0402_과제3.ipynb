# 다양한 마커들을 구글링으로 찾아서 정리!

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# . : 점. point marker
plt.plot([1,2,3],[4,5,6],marker='.',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# o : circle marker, 채워진 원
plt.plot([1,2,3],[4,5,6],marker='o',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# v : triangle_down marker, 채워진 역삼각형
plt.plot([1,2,3],[4,5,6],marker='v',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# ^ : triangle_up marker, 채워진 삼각형
plt.plot([1,2,3],[4,5,6],marker='^',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# < : triangle_left marker, 왼쪽을 향하는 채워진 삼각형
plt.plot([1,2,3],[4,5,6],marker='<',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# > : triangle_right marker, 오른쪽을 향하는 채워진 삼각형
plt.plot([1,2,3],[4,5,6],marker='>',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# 1 : tri_down_ marker, 아래를 향하는 채워진 삼각형틀
plt.plot([1,2,3],[4,5,6],marker='1',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# 2 : tri_up_ marker, 위를 향하는 채워진 삼각형 틀
plt.plot([1,2,3],[4,5,6],marker='2',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# 3 : tri_left_ marker, 왼쪽를 향하는 채워진 삼각형틀
plt.plot([1,2,3],[4,5,6],marker='3',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# 4 : tri_right_ marker, 오른쪽를 향하는 채워진 삼각형틀
plt.plot([1,2,3],[4,5,6],marker='4',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# s : sqare. 채워진 네모
plt.plot([1,2,3],[4,5,6],marker='s',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# p : pentagon 오각형
plt.plot([1,2,3],[4,5,6],marker='p',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```

```python
# * : star
plt.plot([1,2,3],[4,5,6],marker='*',label='mpg.cty',color='red',markersize = 9)
plt.legend()
```
