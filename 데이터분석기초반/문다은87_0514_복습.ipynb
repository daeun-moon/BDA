# 데이터 병합 : merge, concat

# Concat 함수
- df+ df 끼리 합치는 경우 : 두 개의 테이블이 행과 열 기준으로 병합된다.
- axis축 : 좌우로 병합, 상하로 병합 결정 ,ignore_index : 병합후 인덱스를 새롭게 만들지 유무 결정
- concat의 경우 칼럼이 동일한 경우 쉽게 붙일 수있다 
- 시계열 데이터의 경우에는 다를 수 있다

```python
import pandas as pd

df1 = pd.DataFrame({'A':[1,2,3,4],'B':[1,2,3,4]})
df2 = pd.DataFrame({'A':[4,5],'B':[4,5]})
df3 = pd.DataFrame({'A':[5,6],'C':[3,2]})

display(df1)
display(df2)
display(df3)
```

```python
# ignore_index 해서 인덱스 새로 생성
# 행으로 이어져서 병합되는데, 새로운 칼럼의 경우에는 NaN으로 입력됨
pd.concat([df1,df2,df3],ignore_index=True) #axis= 0 디폴트값 : 행을 기준으로 병합된다
```

```python
pd.concat([df1,df3,df2],ignore_index=True) #axis= 0 디폴트값
```

```python
# axis=1로 설정하게 되면, 행으로 병합이 아니라 칼럼이 병합(좌우로 병합)
# 행의 개수는 제일 행이 많은 데이터 프레임 기준, 부족한 값은 다 NaN 처리
pd.concat([df1,df3,df2],axis=1) #axis= 0 디폴트값
```

# merge 함수
- 두 개의 데이터프레임을 join하는 함수 
- 공통된 컬럼을 기준으로 데이터프레임을 병합하는 것이 가능
- 공통된 컬럼이 join key
- 함수 주요 인자
- left : 병합할 첫 번째 데이터 프레임으로 왼쪽 ( 순서가 기준 )
- right : 병합할 두 번째 데이터 프레임으로, 오른쪽 데이터 프레임 
- on : 조인 키로 사용할 컬럼 목록으로, 두 데이터프레임 모두 포함되어야 한다.
- left_on : 왼쪽 데이터프레임에서 조인 키로 사용할 컬럼 목록
- right_on : 오른쪽 데이터프레임에서 조인키로 사용할 컬럼 목록
- left_index : 왼쪽 데이터프레임의 조인 키로 인덱스를 사용하지
- right_idnex: 오른쪽 데이터프레임의 조인키로 인덱스를 사용할지
- how : 조인의 방법을 결정하며, 디폴트는 'inner'

```python
df1 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '영어':[100,50,60,70]})
df2 = pd.DataFrame({'이름':['홍길동','김영희','김철수'],
                   '수학':[50,100,60]})
display(df1)
display(df2)
```

```python
# 디폴트는 inner(교집합)
pd.merge(df1, df2)
```

```python
# on으로 join key 지정
pd.merge(df1, df2 , on='이름')
```

```python
df3 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '수학':[50,100,60,70],
                   '생년월일':['00','01','02','03']})
df4 = pd.DataFrame({'이름':['홍길동','김영희','김철수'],
                   '수학':[50,100,60],
                   '지각횟수':[0,1,2]})
display(df3)
display(df4)
```

```python
# on  사용할때 2개이상의 칼럼도 사용하여 지정가능하다
pd.merge(df3, df4, on = ['이름','수학'])
```

```python
# 이 경우에는 수학칼럼을 공통으로 인식하지 못하고 각각 출력된다
pd.merge(df3,df4, on=['이름'])
```

```python
df3 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '수학':[50,100,60,70],
                   '생년월일':['00','01','02','03']})
df4 = pd.DataFrame({'name':['홍길동','김영희','김철수'],
                   '수학':[50,100,60],
                   '지각횟수':[0,1,2]})
display(df3)
display(df4)
```

```python
# joinkey를 각각의 데이터 프레임에서 지정
# 왼쪽 데이터프레임에서는 이름을, 오른쪽 데이터프레임에서는 name 칼럼을 공통으로 하자
pd.merge(df3,df4, left_on=['이름'],right_on=['name'])
```

```python
# 위 결과에서 수학이 각각으로 인식되니까
# 수학도 공통 칼럼으로 지정 가능
pd.merge(df3,df4, left_on=['이름','수학'],right_on=['name','수학'])
```

```python
df3_sp=df3.set_index('이름')
display(df3_sp)
display(df4)
```

```python
# 위 데이터프레임에서는 이름이 인덱스로, 아래 데이터프레임에서는 이름이 칼럼으로 들어가 있는 상태
# 이때 이름을 공통 속성으로 지정하려면
pd.merge(df3_sp, df4, left_index = True, right_on = 'name')
```

# How
- 조인 키에 대한 inner 기본값
- left :왼쪽 데이터프레임 조인 키를 기준 병합
- right : 오른쪽 데이터프레임의 조인 키를 기준 병합
- outter : 왼쪽과 오른쪽 데이터프레임 조인키에 대한 합집합 기준 

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[1,2,3]})
df2 = pd.DataFrame({'A':[2,3,4], 'C':[7,8,9]})
display(df1)
display(df2)
```

```python
# A 칼럼을 기준으로 교집합을 찾는데, A의 값들이 같은것끼리 매칭시킨다
pd.merge(df1, df2, on = 'A', how = 'inner')
```

```python
# inner가 default이니까 같은 결과가 나온다
pd.merge(df1, df2, on = 'A')
```

```python
# A를 기준으로 하되 왼쪽 데이터프레임 기준으로! 
pd.merge(df1, df2 , on = 'A', how = 'left')
```

```python
# A를 기준으로 하되 오른쪽 데이터프레임 기준으로! 
pd.merge(df1, df2 , on = 'A', how = 'right')
```

```python
# outer는 합집합으로 이해
pd.merge(df1, df2 , on = 'A', how = 'outer')
```

```python
# 실제 데이터 활용
site= pd.read_csv('survey_site.csv')
survey=pd.read_csv('survey_survey.csv')
vit=pd.read_csv('survey_visited.csv')
```

```python
display(survey)
display(site)
display(vit)
```

```python
# site, survey, vit 데이터 병합해보기
# site의 name과 vit의 site를 join 하되, 교집합으로 병합하자
site_vit=pd.merge(site, vit,left_on=['name'],right_on=['site'], how='inner')
site_vit
```

```python
# survey의 taken과 site_vit의 ident를 join하되, 교집합으로 병합
pd.merge(survey,site_vit,left_on=['taken'],right_on=['ident'], how='inner')
```
