```python
import pandas as pd
df = pd.read_csv('spaceship_tt.csv')
```

```python
df['CryoSleep'] = df['CryoSleep']*1
df['Transported'] = df['Transported']*1
df['VIP'] = df['VIP']*1
df.isna().sum()
```

- 결측치를 가장 나이스하게 전처리했는가가 성능에 중요한 기준이 된다.
- 모델링 했는데 성능이 안 나왔다는 것은 전처리를 다시 확인해봐야한다.

# 전처리X 베이스 모델로 정확도 평가

```python
df_base=df[['CryoSleep','Age','VIP','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Transported']]
df_base.dropna(inplace=True)
df_base
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
dt_clf = DecisionTreeClassifier(random_state=111)
X_train, X_test, y_train, y_test = train_test_split(df_base.drop('Transported',axis=1),df_base['Transported'],
                                                   test_size=0.3, random_state=111)
dt_clf = dt_clf.fit(X_train,y_train)
```

```python
X_train
```

```python
#실제 예측값 얼마인지?
from sklearn.metrics import accuracy_score
#하이퍼파라미터 튜닝하지 않은 base_model
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
print(dt_clf.get_params())
```

# My csv 파일로 모델링 진행

```python
mine = pd.read_csv('/content/문다은87_0730_과제2파일.csv')
```

```python
mine['CryoSleep'] =mine['CryoSleep']*1
mine['VIP'] =mine['VIP']*1
```

```python
mine['Transported'] =mine['Transported']*1
```

```python
minedt_clf1 = DecisionTreeClassifier(random_state=111)

X_train, X_test, y_train, y_test = train_test_split(mine.drop('Transported',axis=1),mine['Transported'],
                                                   test_size=0.3, random_state=111)

minedt_clf1 = minedt_clf1.fit(X_train,y_train)

#실제 예측값 얼마인지?
from sklearn.metrics import accuracy_score
#하이퍼파라미터 튜닝하지 않은 base_model
pred1 = minedt_clf1.predict(X_test)
accuracy = accuracy_score(y_test, pred1)
print(accuracy)
print(minedt_clf1.get_params())
```

- 전처리를 안하고도 0.7601102941176471
- 전처리를 했는데 0.7588134135855547 정확도 나온다.
- 성능이 높아지지 않았다!!
- 타겟과의 관계를 고려하고, 실제 데이터들의 시각화를 통해 분포 및 범주들을 확인하고 변수를 추가해야 한다.
- 많은 변수들을 추가한다는게 더 높은 성능을 보이는 것은 아니다.

```python
df['psid_cnt']= df['PassengerId'].str[:4]
```

```python
#전처리를 원하는 건 한 공간에 몇 명이 같이 탑승했는가?
df_psid=pd.DataFrame(df.groupby('psid_cnt')['PassengerId'].count()).reset_index()
```

```python
df_psid
```

```python
df_pre1 =pd.merge(df, df_psid, left_on= 'psid_cnt',right_on='psid_cnt', how='inner')
```

```python
df_pre1.PassengerId_y.value_counts()
```

```python
df_pre1['PassengerId_y'].isna().sum()
```

```python
df_pre1.columns
```

```python
df_pre1=df_pre1[['HomePlanet', 'CryoSleep', 'Cabin', 'Destination',
       'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa',
       'VRDeck', 'Name', 'Transported', 'PassengerId_y']]
```

```python
df_pre1=df_pre1[['CryoSleep','Age','VIP','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Transported','PassengerId_y']]
```

```python
df_pre1
```

```python
df_pre1.dropna(inplace=True)
```

```python
dt_clf6 = DecisionTreeClassifier(random_state=111)

X_train, X_test, y_train, y_test = train_test_split(df_pre1.drop('Transported',axis=1),df_pre1['Transported'],
                                                   test_size=0.3, random_state=111)

dt_clf6 = dt_clf6.fit(X_train,y_train)

#실제 예측값 얼마인지?
from sklearn.metrics import accuracy_score
#하이퍼파라미터 튜닝하지 않은 base_model
pred6 = dt_clf6.predict(X_test)
accuracy = accuracy_score(y_test, pred6)
print(accuracy)
print(dt_clf6.get_params())
```
