# 시각화 
- 시각화를 하는 이유는 패턴을 찾고, 경향을 보고, 분석을 시작 및 설명하기 위해서
- 나의 주장을 뒤받침 하는 용도
- 분석 후 분석에 대한 근거를 제시할 때 데이터 분석을 잘 모르는 사람들에게 시각화를 통해 더 쉽고 빠르게 이해 시킬 수 있다. 
- 시각화를 통해서 무엇을 보여주고 싶은지를 명확히 해야한다. 
- 산점도, 히스토그램, 막대그래프, 박스플랏 등이 있다

```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt 
%matplotlib inline # 매직커맨드 show 함수 하지 않아도 바로 그래프 보여준다
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.show()
```

```python
# 여러개 그리면 자동으로 합쳐진다
plt.scatter([1,2,3],[4,5,6])
plt.scatter([5,6,7],[10,5,6])
plt.scatter([2,4,8],[1,2,5])
plt.show()
```

```python
# 그래프의 축과 이름 설정하기
plt.scatter([1,2,3],[4,5,6])
plt.title('mpg') # 제목
plt.xlabel('mpg.hwy') # x축에 대한 설명
plt.ylabel('mpg.cty') # y축에 대한 설명
```

```python
# 축의 범위 설정 : xlim, ylim 활용
plt.scatter([1,2,3],[4,5,6])
plt.xlim(0,200)
plt.ylim(30, 50)
```

```python
# 눈금값, 범례 조절
plt.scatter([1,2,3],[4,5,6])
plt.xticks([1,2,3], ['A',"B",'C']) #x축 눈금값 지정
```

```python
# 범례 : legned, label을 통해 범례 값 지정
plt.scatter([1,2,3],[4,5,6],label='mpg.cty')
plt.scatter([6,7,8],[3,2,5],label='mpg.hwy')
plt.legend()
```

```python
# 범례 위치 지정 
plt.scatter([1,2,3],[4,5,6],label='mpg.cty')
plt.scatter([6,7,8],[3,2,5],label='mpg.hwy')
plt.legend(loc = 'center')
```

```python
# 그래프 크기 조절
## 도화지 사이즈 지정으로 크기 조절 가능
## figure 함수로 figsize 도화지 크기 지정해준다. 
plt.figure(figsize =(10,5))
plt.scatter([1,2,3],[4,5,6],label='mpg.cty')
```

```python
import seaborn as sns
sns.set() # 기본값은 grid
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style = 'whitegrid')
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style = 'dark')
plt.scatter([1,2,3],[4,5,6])
```

```python
# sns.set(style = 'a') #기본값이 grid
# plt.scatter([1,2,3],[4,5,6]) #그래프가 바뀌어진다.
# 실행하면 오류 뜸. 오류 확인하여 가능한 style 확인 가능
```

```python
# 다양한 그래프 그려보기
sns.set(style='whitegrid')
plt.plot([1,2,3],[4,5,6],label = 'mpg.cty')
plt.plot([6,7,8],[3,2,5], label = 'mpg.hwy')
plt.legend()
```

```python
# 다양한 색상 설정
sns.set(style = 'whitegrid')
plt.plot([1,2,3],[4,5,6],label = 'mpg.cty',color = 'red')
plt.plot([6,7,8],[3,2,5],label = 'mpg.hwy',color ='blue')
plt.legend()
```

```python
# 선 스타일과 두께 설정
sns.set(style='whitegrid')
plt.plot([1,2,3],[4,5,6],label='mpg.cty',color='red',linestyle='dashed', linewidth=10)
plt.plot([6,7,8],[3,2,5],label='mpg.hwy',color='blue',linestyle='dotted', linewidth=5)
plt.legend()
```

```python
# 마커설정
# 값에 대한 포인트를 지정해서 해당 값이 어디에 위치하는지 찍어준다
# . o * 등이 있다. 
plt.plot([1,2,3],[4,5,6],marker='s', label = 'mpg.cty',color='red',linestyle='dashed')
plt.plot([6,7,8],[3,2,5],marker='*',label = 'mpg.hwy',color = 'blue',linestyle='dotted')
```

```python
plt.plot([1,2,3],[4,5,6],marker='s',markersize =10,  label = 'mpg.cty',color='red',linestyle='dashed')
plt.plot([6,7,8],[3,2,5],marker='*',label = 'mpg.hwy',color = 'blue',linestyle='dotted')
```
