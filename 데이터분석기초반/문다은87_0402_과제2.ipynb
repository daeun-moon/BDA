### 필수과제 2
- 범례 값 만들어서 비교해 보기!
- left
- center
- right
- upper left , right, center
- lower left, right, center


```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
#자동으로 계산하여 최적의 위치에 범례를 위치시킨다. 
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='best')
```

```python
# 위쪽 상단에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='upper right')
```

```python
# 왼쪽 상단에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='upper left')
```

```python
# 왼쪽 하단에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='lower left')
```

```python
# 오른쪽 하단에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='lower right')
```

```python
# 중앙 오른쪽에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='right')
```

```python
# 중앙 왼쪽에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='center left')
```

```python
# 중앙 오른쪽에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='center right')
```

```python
# 정중앙에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='center')
```

```python
# 중앙 아랫쪽에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='lower center')
```

```python
# 중앙 윗쪽에 범례 위치
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([6,7,8],[3,2,5],label='y')
plt.legend(loc='upper center')
```
