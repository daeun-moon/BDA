```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

```python
#1 embarked 별 생존확률 보여주는 새로운 변수 만들기
titanic.groupby('embarked').agg(sur_pro=('survived','mean'))
```

```python
#2 sex,pclass 별 생존확률 보여주는 새로운 변수 만들기
titanic.groupby(['sex','class']).agg(sur_pro=('survived','mean'))
```

```python
#3 alone 유무에 따른 생존확률 보여주는 새로운 변수 만들기
titanic.groupby('alone').agg(sur_pro=('survived','mean'))
```

```python
#4 age로 내림차순 정렬한 후 survived와 age만 보기
titanic.sort_values('age', ascending = False)[['survived','age']]
```

```python
# 5 age가 30이하인 경우에 평균 생존확률 새로운 변수로 확인
titanic.query('age<=30').agg(pro_sur=('survived','mean'))
```

```python
#6 fare가 0인 경우에 평균 생존확률 새로운 변수로 확인
titanic.query('fare==0').agg(pro_sur=('survived','mean'))
```

```python
#7 pclass 별 평균 fare 값과 데이터 개수 보여주는 새로운 변수를 통해 확인
titanic.groupby('pclass').agg(mean_fare = ('fare','mean'), n_data = ('fare','count'))
```

```python
#8 'sibsp', 'parch' 변수 합쳐서 familysize라는 새로운 변수 만들기, familysize, sibsp, parch,survive만 조회
titanic = titanic.assign(familysize= titanic['sibsp']+titanic['parch'])[['familysize','sibsp','parch','survived']]
titanic
```

```python
#9 위에서 만든 familysize 활용하여 familysize 별 생존확률 확인+ 각 데이터 개수 구하기
titanic.groupby('familysize').agg(sur_pro = ('survived','mean'), n=('survived','count'))
```

```python
#10 familysize와 survived만 선택하고 familysize로 내림차순 정렬하
titanic[['familysize','survived']].sort_values('familysize', ascending = False)
```
